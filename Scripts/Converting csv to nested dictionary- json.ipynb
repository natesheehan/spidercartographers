```python
import pandas as pd
import csv
```

```python
data = pd.read_csv('Data/msoa_flows.csv')
```

```python
data.columns
```

```python
# melt the data so that you have the variable names as column values not just headers 

melted = pd.melt(data, 
                 id_vars =['Area.of.residence'], 
                 value_vars =['Area.of.workplace', 'All.categories..Method.of.travel.to.work', 'Work.mainly.at.or.from.home', 
                     'Underground..metro..light.rail..tram', 'Train', 'Bus..minibus.or.coach', 'Taxi',
                     'Motorcycle..scooter.or.moped', 'Driving.a.car.or.van',
                     'Passenger.in.a.car.or.van', 'Bicycle', 'On.foot',
                     'Other.method.of.travel.to.work', 'Origin name', 'Orig_Lon', 'Orig_Lat',
                     'Destination name', 'Dest_Lon', 'Dest_Lat']) 

melted.head(20)
```

```python
len(melted)
```

```python
# long ass dataframe - subset to check if below code works
melted = melted.head(50000)
```

```python
# convert to csv for the next function 
melted.to_csv('csv_for_json.csv')
```

```python
# from: https://stackoverflow.com/questions/25485681/creating-a-nested-dictionary-from-a-csv-file-with-python

import csv

new_data_dict = {}
with open("csv_for_json.csv", 'r') as data_file:
    data = csv.DictReader(data_file, delimiter=",")
    for row in data:
        item = new_data_dict.get(row["Area.of.residence"], dict())
        item[row["variable"]] = row["value"]

        new_data_dict[row["Area.of.residence"]] = item

print (new_data_dict)
```

```python
# convert to JSON
import json

json_output = json.dumps(new_data_dict)
#print(json_output)
```

```python
# save to JSON file
with open('Data/flows.json', 'w') as f:
    json.dump(json_output, f)
```

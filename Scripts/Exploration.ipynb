```python
#before transforming drop the MSOA column
#create a new dataframe so retain the original MSOA column
cleaned1 = cleaned.drop(["MSOA"], axis = 1 )
```

```python
#cleaning them by log transformation

import numpy as np

for column in cleaned1.columns:
    try:
        cleaned1[column] = np.log10(cleaned1[column])
    except (ValueError, AttributeError):
        pass
```

```python
#log does not work with 0's for which there were many
#therefore replace any infinity values with 0 again afterwards
cleaned1 = cleaned1.replace([np.inf, -np.inf], 0)
```

```python
#plot the resulting distributions using the same as before
f, axes = plt.subplots(4,5, figsize = (20,20), sharex=False)


sns.distplot(cleaned1["Bus_stops"], ax=axes[0,0])
sns.distplot(cleaned1["Train_stations"], ax=axes[0,1])
sns.distplot(cleaned1["metro_stations"], ax=axes[0,2])
sns.distplot(cleaned1["HH_owning_cars_perc"], ax=axes[0,3])
sns.distplot(cleaned1["work_from_home_perc"], ax=axes[0,4])
sns.distplot(cleaned1["underground_metro_perc"], ax=axes[1,0])
sns.distplot(cleaned1["train_perc"], ax=axes[1,1])
sns.distplot(cleaned1["bus_perc"], ax=axes[1,2])
sns.distplot(cleaned1["taxi_perc"], ax=axes[1,4])
sns.distplot(cleaned1["motorcycle_perc"], ax=axes[2,0])
sns.distplot(cleaned1["bicycle_perc"], ax=axes[2,1])
sns.distplot(cleaned1["on_foot_perc"], ax=axes[2,2])
sns.distplot(cleaned1["other_perc"], ax=axes[2,3])
sns.distplot(cleaned1["avg_time_car"], ax=axes[2,4])
sns.distplot(cleaned1["avg_time_bus"], ax=axes[3,0])
sns.distplot(cleaned1["avg_time_rail"], ax=axes[3,1])
sns.distplot(cleaned1["car_perc"], ax=axes[3,2])
```

```python
#add the MSOA value back in
cleaned1["MSOA"] = cleaned["MSOA"]
```

```python
#check to see what the outcome looks like
cleaned1
```

```python
#output the results to a csv
cleaned1.to_csv("Data/transport_log_transform.csv", index=False)
```

```python
#running the next stage which is the standardisation
#drop the MSOA value again

cleaned1.drop(["MSOA"], axis = 1, inplace = True)
```

```python
#get the columns names to then re-add to the numpy array
column_names = list(cleaned1.columns)
```

```python
#use the preprocessing function which does a normal standardisation


import sklearn.preprocessing as preprocessing

#this takes arguments: X ,axis = 0, with_mean = True, with_std = True, copy = True
#X is the data to centre and scale
#axs: int (0 by default) - axis used to compute means and Std along, if 0  independentky
#standardise each feature, other (if 1) standardize each sample
#with_mean: boolean, True by default, if true centre the data before scaling
#with_std: boolean, True by default, if True scale the data to unit variance
#copy: boolean, optional, default True, set to False to perform inplace row normalisation and avoid a copy

#NaNs are treated as missing values: disregarded

scaled = preprocessing.scale(cleaned1) # add your dataframe name here

#this creates a numpy array
scaled
```

```python
#transform the numpy array back into a dataframe
log_then_normal = pd.DataFrame(scaled)
```

```python
#add back the columns names stored before

log_then_normal.columns = column_names

```

```python
#add back the MSOA columns
log_then_normal["MSOA"] = cleaned["MSOA"]
```

```python
#write the resulting output to a csv
log_then_normal.to_csv("Data/transport_log_then_normal.csv", index=False)
```

## range standardisation

```python
#transform df to np array
cleaned_array = cleaned1.values

#range standardisation
min_max_scaler = preprocessing.MinMaxScaler()
min_max_scaled = min_max_scaler.fit_transform(cleaned_array)
# transform to dataframe
range_standardized = pd.DataFrame(min_max_scaled, columns=column_names)
range_standardized
```

```python
#then add the MSOA column
range_standardized["MSOA"] = cleaned["MSOA"]

range_standardized
```

```python
range_standardized.to_csv("Data/transport_log_then_range.csv", index=False)
```

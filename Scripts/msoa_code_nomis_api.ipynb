```python
import pandas as pd
```

```python
## From NOMIS...this only has 7194 MSOA codes, but we want 7201 https://geoportal.statistics.gov.uk/datasets/middle-layer-super-output-areas-december-2001-names-and-codes-in-england-and-wales/data
#msoa_codes = pd.read_csv("Data/Middle_Layer_Super_Output_Areas__December_2001__Names_and_Codes_in_England_and_Wales.csv")
#msoa_codes               
```

```python
# read in an old dataset we have
msoa_codes = pd.read_csv("Data/MSOA households.csv")
# keeep only necessary columns
msoa_codes = msoa_codes[['MSOA Code', 'MSOA Name']]
msoa_codes
```

```python
# read in names and nomis codes from https://www.nomisweb.co.uk/api/v01/dataset/NM_1208_1/USUAL_RESIDENCE.def.htm
msoa_nomis = pd.read_csv("Data/nomis_msoa_names_codes.csv")
msoa_nomis
```

```python
# join the two tables
msoa_matched = pd.merge(msoa_codes,
                           msoa_nomis,
                           left_on = "MSOA Name",
                           right_on = "Description",
                          how = 'left')

msoa_matched
```

```python
msoa_matched.drop(columns=['Description'], inplace = True)
msoa_matched
```

```python
# save as csv
msoa_matched.to_csv("Data/msoa_code_nomis_api.csv", index=False)
```

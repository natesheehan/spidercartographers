## From: https://www.nomisweb.co.uk/census/2011/quick_statistics

## Employment Category 

```python
employment2011 = pd.read_csv('Data/Socioeconomic_Data/employment_category.csv')
employment2011
```

```python
# Check what values we can use
employment2011.columns
```

```python
# get economic activity as % of 'All usual residents aged 16 to 74'

employment2011.iloc[:,5:] = employment2011.iloc[:,5:].div(employment2011['Sex: All persons; Economic Activity: All usual residents aged 16 to 74; measures: Value'], 
                                                                axis = 0).multiply(100).round(1)
employment2011
```

```python
employment2011 = employment2011[['geography code', 'Sex: All persons; Economic Activity: Economically active: Unemployed; measures: Value']]
employment2011.rename(columns={"Sex: All persons; Economic Activity: Economically active: Unemployed; measures: Value": "percent_unemployed"}, inplace = True)

employment2011
```

## Qualification Level

```python
qualification2011 = pd.read_csv('Data/Socioeconomic_Data/qualification_level.csv')
qualification2011
```

```python
# convert the Qualification figures to percentages using the all categories column as the denominator
qualification2011.iloc[:,5:] = qualification2011.iloc[:,5:].div(qualification2011['Qualifications: All categories: Highest level of qualification; measures: Value'], 
                                                                axis = 0).multiply(100).round(1)
qualification2011

#qualification2011.select_dtypes(np.number).div(qualification2011['Qualifications: All categories: Highest level of qualification; measures: Value'], 
#                      axis = 0).multiply(100).round(1)
```

```python
qualification2011.columns
```

```python
# take only the values for qualification levels (1, 2, 3, 4 etc)

qualification2011v1 = qualification2011[['geography code','Qualifications: No qualifications; measures: Value',
       'Qualifications: Highest level of qualification: Level 1 qualifications; measures: Value',
       'Qualifications: Highest level of qualification: Level 2 qualifications; measures: Value',
       'Qualifications: Highest level of qualification: Level 3 qualifications; measures: Value',
       'Qualifications: Highest level of qualification: Level 4 qualifications and above; measures: Value']]
```

```python
# CALCULATE A WEIGHTED AVERAGE OF AVERAGE QUALIFICATION LEVEL 

# get row sums. They do not add up to 100% as we have removed some columns
qualification2011v1['sum'] = qualification2011v1.sum(axis=1)
#qualification2011v1
```

```python
# get a weighted average of qualification/education level

qualification2011v1['avg_qual_level'] = (
(qualification2011v1.iloc[:,1] / qualification2011v1['sum'])* 0 + 
(qualification2011v1.iloc[:,2] / qualification2011v1['sum'])* 1 + 
(qualification2011v1.iloc[:,3] / qualification2011v1['sum'])* 2 +
(qualification2011v1.iloc[:,4] / qualification2011v1['sum'])* 3 +
(qualification2011v1.iloc[:,5] / qualification2011v1['sum'])* 4
).round(1)

#qualification2011v1
```

```python
# subset the required columns
qualification2011v1 = qualification2011v1[['geography code', 'avg_qual_level']]
qualification2011v1
```

 Alternatively, just get the % of people with Level 4 or above

```python
qualification2011v2 = qualification2011[['geography code',
                                         'Qualifications: Highest level of qualification: Level 4 qualifications and above; measures: Value']]

qualification2011v2.rename(columns={"Qualifications: Highest level of qualification: Level 4 qualifications and above; measures: Value": "percent_at_or_above_qual_level_4"}, inplace = True)
```

## Population Density 
1. MSOA Areas: 
http://geoportal.statistics.gov.uk/datasets/5a94044d113a4bd5bd895975d6612b05
2. MSOA Populations: https://www.ons.gov.uk/peoplepopulationandcommunity/populationandmigration/populationestimates/datasets/middlesuperoutputareamidyearpopulationestimates

1. MSOA Areas

```python
msoa_area = pd.read_csv('Data/Socioeconomic_Data/MSOA_Areas_2011_Census/SAM_MSOA_DEC_2011_EW.csv',  encoding = "ISO-8859-1")
msoa_area
```

#### We have 4 Areas
- Total Extent (AREAEHECT)
- Area to Mean High Water (clipped to the coastline) (AREACHECT)
- Area of Inland Water (AREAIHECT)
- Area to Mean High Water Excluding Area of Inland Water (Land Area) (AREALHECT)

We will use the last one as it excludes all water bodies

```python
msoa_area = msoa_area[['MSOA11CD', 'AREALHECT']]
msoa_area
```

2. MSOA Populations

```python
msoa_pop2018 = pd.read_csv('Data/Socioeconomic_Data/population_2018.csv', skiprows = 4)
msoa_pop2018 
```

```python
# drop the aggregate rows (these are summations of MSOAs in the same County)
msoa_pop2018  = msoa_pop2018 [msoa_pop2018 ['MSOA'].notna()]
```

```python
msoa_pop2018 = msoa_pop2018[['Area Codes', 'All Ages']]
```

3. Calculate Population Densities

```python
# merge the two dataframes

pop_density = pd.merge(msoa_area, msoa_pop2018, how = 'left', left_on='MSOA11CD', right_on='Area Codes')
pop_density.drop('Area Codes', axis=1, inplace=True)
#pop_density
```

```python
pop_density.dtypes
```

```python
#turn 'All Ages' column to numeric, from: https://www.reddit.com/r/learnpython/comments/5ktuhv/how_do_i_remove_commas_from_data_frame_column/

pop_density['All Ages'] = pop_density['All Ages'].str.replace(",","").astype(float)
#pop_density
```

```python
# Get Population Density
pop_density['Pop_Per_Hectare'] = (pop_density['All Ages'] / pop_density['AREALHECT']).round(1)
# keep only the desired columns
pop_density = pop_density[['MSOA11CD', 'Pop_Per_Hectare']]
```

# Merge the Data

```python
merged1 = pd.merge(income2018, employment2011, how = 'left', left_on='MSOA code', right_on='geography code')
#merged1
```

```python
merged2 = pd.merge(merged1, pop_density, how = 'left', left_on='MSOA code', right_on='MSOA11CD')
#merged2
```

```python
merged3 = pd.merge(merged2, qualification2011v2, how = 'left', left_on='MSOA code', right_on='geography code')
merged3
```

```python
# import cluster labels csv. This is log-zscore-kmeans
clusters = pd.read_csv('Data/Clustering_Results/clusterlabels_cleaneddata3.csv')
#clusters
```

```python
# Extract relevant columns
clusters = clusters[['MSOA', 'log_zscore_kmeans_cluster']]
clusters
```

```python
# merge clustering results
merged4 = pd.merge(merged3, clusters, how = 'left', left_on='MSOA code', right_on='MSOA')
merged4
```

```python
# Drop unnecessary columns
merged4.drop(['geography code_x', 'MSOA11CD', 'geography code_y', 'MSOA'], axis=1, inplace=True)
```

```python
#save to csv
merged4.to_csv('Data/Socioeconomic_Data/socioeconomic_all.csv',  encoding='utf-8')
```

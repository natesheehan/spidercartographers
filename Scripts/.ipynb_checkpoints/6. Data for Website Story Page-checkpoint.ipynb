### Add the Urban / Rural Categorical Data 

(from data.gov https://data.gov.uk/dataset/9c0e093d-d267-4eb8-90d8-54475ab4d1ff/rural-urban-classification-2011-of-middle-layer-super-output-areas-in-england-and-wales)

```python
urban = pd.read_csv("Data/Rural_Urban/Rural_Urban_Classification_(2011)_of_Middle_Layer_Super_Output_Areas_in_England_and_Wales.csv")
urban
```

Get unique rural/urban classifications

```python
urban.RUC11.unique()
```

It is very detailed. For our purposes, we will settle for Urban OR Rural. The function below adds an is_urban binary variable and assigns a value of 1 to all rows with the word Urban in the RUC11 column

```python
urban["is_urban"] = urban["RUC11"].map(lambda x: 1 if "Urban" in x else 0)
urban
```

```python
urban = urban[['MSOA11CD','is_urban']]
urban
```

## Choropleth Maps

```python
# all variables
map_data = all_data
map_data
```

```python
# get MSOA areas to plot densities (e.g. Bus stops/hectare)
msoa_area = pd.read_csv('Data/Socioeconomic_Data/MSOA_Areas_2011_Census/SAM_MSOA_DEC_2011_EW.csv',  encoding = "ISO-8859-1")
msoa_area
```

#### We have 4 Areas
- Total Extent (AREAEHECT)
- Area to Mean High Water (clipped to the coastline) (AREACHECT)
- Area of Inland Water (AREAIHECT)
- Area to Mean High Water Excluding Area of Inland Water (Land Area) (AREALHECT)

We will use the last one as it excludes all water bodies

```python
# keep only necessary columns
msoa_area = msoa_area[['MSOA11CD', 'AREALHECT']]
msoa_area
```

```python
# merge the two dataframes
map_data = pd.merge(map_data, msoa_area, how = 'left', left_on='MSOA', right_on='MSOA11CD')
map_data.drop('MSOA11CD', axis=1, inplace=True)
map_data
```

```python
# add columns to get density of PT nodes
map_data['bus_stop_per_km2']       = map_data['Bus_stops'] / (map_data['AREALHECT'] / 100) 
map_data['metro_stations_per_km2'] = map_data['metro_stations'] / (map_data['AREALHECT'] / 100) 
map_data['train_stations_per_km2'] = map_data['Train_stations'] /(map_data['AREALHECT'] / 100) 
```

```python
import geopandas as gpd
```

```python
# read in the geographic boundaries
geo_data = gpd.read_file('Data/MSOA_shp/Middle_Layer_Super_Output_Areas__December_2011__Boundaries.shp')
geo_data
```

```python
geo_data.columns
```

```python
# remove unnecessary columns 
geo_data.drop(['objectid', 'msoa11nm','msoa11nmw','st_area(sh', 'st_length('], axis=1, inplace=True)
```

```python
# merge dataframe with spatial data for plotting
for_map = pd.merge(geo_data, map_data, how = "left", left_on='msoa11cd', right_on='MSOA')

for_map.drop('msoa11cd', axis=1, inplace=True)

# try keeping necessary columns only 
#for_map = for_map[['MSOA', 'TCITY15NM', 'bus_stop_per_hect', 'geometry']]
for_map
```

```python
# create data subsets to plot cities independantly (using bool column done previously)
london_data = for_map[is_london]
birmingham_data = for_map[is_birmingham]
blackpool_data = for_map[is_blackpool]
liverpool_data = for_map[is_liverpool]
```

```python
#london_data['bus_stop_per_km2'].describe()
blackpool_data['bus_stop_per_km2'].describe()
```

```python
# bokeh for choropleth maps

import json
from bokeh.io import show, output_notebook
from bokeh.models import (CDSView, ColorBar, ColumnDataSource,
                          CustomJS, CustomJSFilter, 
                          GeoJSONDataSource, HoverTool,
                          LinearColorMapper, Slider)
from bokeh.layouts import column, row, widgetbox
from bokeh.palettes import brewer
from bokeh.plotting import figure
from bokeh.io import export_png
```

```python
# Input GeoJSON source that contains features for plotting
#geosource = GeoJSONDataSource(geojson = london_data.to_json())
geosource = GeoJSONDataSource(geojson = blackpool_data.to_json())
```

```python
#Define a sequential multi-hue color palette.
palette = brewer['Greens'][8]
#Reverse color order so that dark blue is highest obesity.
palette = palette[::-1]
#Instantiate LinearColorMapper that linearly maps numbers in a range, into a sequence of colors.
color_mapper = LinearColorMapper(palette = palette, low = 0, high = 100)
#Define custom tick labels for color bar.
#tick_labels = {'0': '0%', '5': '5%', '10':'10%', '15':'15%', '20':'20%', '25':'25%', '30':'30%','35':'35%', '40': '>40%'}

#Create color bar. 
color_bar = ColorBar(color_mapper=color_mapper, label_standoff=8, width = 500, height = 20,
#border_line_color=None,location = (0,0), orientation = 'horizontal', major_label_overrides = tick_labels)
border_line_color=None, location = (0,0), orientation = 'horizontal')
```

```python
#Create figure object.
p = figure(title = 'Bus Stops / km2', plot_height = 600 , plot_width = 950, toolbar_location = None)
#p.xgrid.grid_line_color = None
#p.ygrid.grid_line_color = None
p.axis.visible = False
#Add patch renderer to figure. 
p.patches('xs','ys', source = geosource, fill_color = {'field' :'bus_stop_per_km2', 'transform' : color_mapper},
          line_color = 'black', line_width = 0.25, fill_alpha = 1)
#Specify figure layout.
p.add_layout(color_bar, 'below')
#Display figure inline in Jupyter Notebook.
#output_notebook()
```

```python
#Display figure inline in Jupyter Notebook.
output_notebook()
```

```python
#Display figure.
show(p)
```

```python
# save as png
export_png(p, filename="Plots/blackpool_bus_stop_density.png")
```

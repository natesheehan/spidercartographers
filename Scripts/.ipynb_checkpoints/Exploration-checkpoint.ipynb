```python
import seaborn as sns

columns = list(cleaned.columns)

columns.remove("MSOA")

for i in columns:
    sns.distplot(cleaned[i], hist=True, kde=False, 
                 bins=int(180/5),
                 hist_kws={'edgecolor':'black'})
```

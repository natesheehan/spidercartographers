```python
#get the all transport data csv
all_data = pd.read_csv("Data/Clustering_Results/clusterlabels_cleaneddata.csv")
#check the resulting csv
all_data
```

```python
#get all the column values
all_data.columns
```

```python
#store the columns that we want to remove in a list
columns = ['Unnamed: 0','yj_idr_Kmeans_cluster', 'yj_DB_cluster', 'log_zscore_DB_cluster',
       'log_zscore_kmeans_cluster', 'log_zscore_H_cluster',
       'log_range_DB_cluster', 'log_range_H_cluster',
       'log_range_kmeans_cluster', 'yj_idr_DB_cluster', 'yj_idr_H_cluster',
       'yj_H_cluster', 'yj_Kmeans_cluster', 'yj_Zscore_DB_cluster',
       'yj_Zscore_H_cluster', 'yj_zscore_Kmeans_cluster', 'log_idr_H_cluster',
       'log_idr_DB_cluster', 'log_idr_kmeans_cluster']
#drop all the columns that we are not interested in inplace
all_data.drop(columns=columns, inplace=True)
```

```python
#check to see which columns contain any NaN values
all_data.isna().any()
```

```python
#for columns where NaN should represent a 0 replace the NaN with a o
na_list = ["Bus_stops", "Train_stations", "metro_stations"]
for x in na_list:
    all_data[x].replace(np.nan, 0, inplace= True) 
```

```python
#then look at the resulting descriptions of each value
msoa_described = all_data.describe()
msoa_described
```

```python
#counting the number of MSOAs with train, bus and metro stations
all_data.isin([0]).sum()
```

```python
#getting the total number of bus stops, metro stations and train stations
bus_stations = all_data["Bus_stops"].sum()
train_stations = all_data["Train_stations"].sum()
metro_stations = all_data["metro_stations"].sum()
print(bus_stations, train_stations, metro_stations)
```

```python
#store those values in a resulting csv
msoa_described.to_csv("Data/all_transport_averages.csv")
```
